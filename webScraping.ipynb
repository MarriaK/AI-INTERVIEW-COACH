## Merge all json files

```python
import glob

# Get a list of all JSON files in the current directory
json_files = glob.glob("*.json")

# Initialize an empty list to store the merged data
merged_data = []

# Loop through each JSON file and append its content to the merged_data list
for file in json_files:
    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)
        merged_data.extend(data)

# Save the merged data into a new JSON file
output_file = "merged_data.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=4)

print(f"Merged {len(json_files)} JSON files into {output_file}")
```
